## Debugging: The Dangers of Glob!

Team debugging with Daniel Kennedy helped reveal that the triming clause to remove the 0000th file [1:] was inside of sorted.
This lead to the splicing to not occur on the file we expected it to because glob lists files based on timestamp, rather than alphanumeric 
So glob literally globs together the files in it's own order, that's why we add `sorted()`, but we need to careful where we place the [1:]
